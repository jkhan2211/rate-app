```python
# Your API Key
api_key = os.getenv("API_KEY")  # Ensure this environment variable is set correctly

# Location (latitude, longitude) and radius (in meters)
location = "43.6532,-79.3832"  # Example location (Toronto, Canada)
radius = 10000  # 10 km radius

# Google Places Nearby Search API endpoint
url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={location}&radius={radius}&type=restaurant&key={api_key}&keyword=food"

# Send the request to the API
response = requests.get(url)
places_data = response.json()

# List of specific cuisines to filter
specific_cuisines = {
    'Italian': ['italian'],
    'Japanese': ['japanese'],
    'Indian': ['indian'],
    'Mexican': ['mexican'],
    'Chinese': ['chinese'],
    'French': ['french'],
    'Vegan': ['vegan'],
    'Thai': ['thai'],
    'Spanish': ['spanish'],
    'American': ['american'],
    'British': ['british'],
    'Deli': ['deli']
}

# Check if the request was successful
if response.status_code == 200:
    print("Request Successful")
    if 'results' in places_data and places_data['results']:
        # Output the results
        for place in places_data['results']:
            name = place.get('name')
            rating = place.get('rating', 'N/A')  # Default to 'N/A' if rating is not available
            types = place.get('types', [])  # List of types from API
            # Address is optional; some results may not have it
            address = place.get('vicinity', 'No address available')  # Use 'vicinity' as address

            # Determine the cuisine based on the types and name
            cuisine = 'Unknown'
            for specific_cuisine, keywords in specific_cuisines.items():
                if any(keyword in ' '.join(types).lower() for keyword in keywords) or any(keyword in name.lower() for keyword in keywords):
                    cuisine = specific_cuisine
                    break

            print(f"Name: {name}")
            print(f"Rating: {rating}")
            print(f"Cuisine: {cuisine}")
            print("-" * 40)  # Separator line for readability
    else:
        print("No restaurants found within the specified radius.")
else:
    print("Request Failed")
    print("Status Code:", response.status_code)
    print("Error Message:", places_data.get('error_message'))
```
